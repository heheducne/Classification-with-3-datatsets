```python
from tokenize import PlainToken
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
```

```python
dataset=pd.read_csv('Salary_Data.csv')
z=dataset.values
x=dataset.iloc[:,0].values.reshape(-1,1)
y=dataset.iloc[:,1].values.reshape(-1,1)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

Linear Regression

```python
print('LR init')
linear = LinearRegression()
linear.fit(x_train,y_train)

y_pred=linear.predict(x_test)#predict

print(f"Accuracy = {linear.score(x_test, y_test)}")
print(f"Mean square error = {mean_squared_error(y_test, y_pred)}")

plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_test,y_pred,color='blue')
plt.show()
```

Decision Tree

```python
print('Decision init')
dt = DecisionTreeRegressor()

dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)

print(f"Accuracy = {dt.score(x_test, y_test)}")
print(f"Mean square error = {mean_squared_error(y_test, y_pred)}")

plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_test,y_pred,color='blue')
plt.show()
```

Random Forest

```python
print('RF init')
rf = RandomForestRegressor()

rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)

print(f"Accuracy = {rf.score(x_test, y_test)}")
print(f"Mean square error = {mean_squared_error(y_test, y_pred)}")

plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_test,y_pred,color='blue')
plt.show()
```

SVR

```python
print('SVR init')
svr =SVR(kernel ='poly', C=1000)

svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)

print(f"Accuracy = {svr.score(x_test, y_test)}")
print(f"Mean square error = {mean_squared_error(y_test, y_pred)}")

plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_test,y_pred,color='blue')
plt.show()
```

SVR with Polynominal kernel

```python
print('SVR init')
sv =  SVR(kernel="poly", C=100, degree=3, epsilon=0.1, coef0=1)
sv.fit(x_train, y_train)

y_pred = sv.predict(x_test)

plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_test,y_pred,color='blue')
plt.show()
```

Polynomial

```python
print('LR init')
degree=6

model = make_pipeline(PolynomialFeatures(degree),LinearRegression())
model.fit(x_train, y_train)

x_seq = np.linspace(x.min(),x.max(),300).reshape(-1,1)
y_pred = model.predict(x_seq)

print(f'score: {model.score(x_test, y_test)}')

plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_seq, y_pred, color='blue')
plt.show()
```
